```python
n = 10
df['Momentum'] = df['VNI_Close price'].diff(n)
```

```python
df['Inflation'] = 100 * (df['CPI'].diff() / df['CPI'].shift(1))
```

```python
df.head()
```

```python
df = df.bfill()
```

```python
# Tính T_spread_US và T_spread_VN với tên cột chính xác
df['T_spread_US'] = df['US10_Price'] - df['US2_Price']
df['T_spread_VN'] = df['VN10_Price'] - df['VN2_Price']

# Bỏ 4 cột không cần thiết
df.drop(columns=['US10_Price', 'US2_Price', 'VN10_Price', 'VN2_Price'], inplace=True)


```

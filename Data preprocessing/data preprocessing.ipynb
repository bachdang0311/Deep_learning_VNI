```python
# Cách 1: Giảm độ lệch chuẩn
std_min_adj, std_max_adj = 0.05, 0.2  # Giảm độ lệch chuẩn

daily_data_1 = []
for index, row in CPI.iterrows():
    month_mean = row["CPI (%)"]
    date = pd.to_datetime(row["Date"])
    days_in_month = pd.Period(date, freq='M').days_in_month
    std_dev = np.random.uniform(std_min_adj, std_max_adj)
    daily_values = np.random.normal(loc=month_mean, scale=std_dev, size=days_in_month)
    daily_values = np.clip(daily_values, 0, None)  # Đảm bảo không âm
    dates = pd.date_range(start=date, periods=days_in_month, freq='D')
    daily_data_1.extend(zip(dates, daily_values))

CPI = pd.DataFrame(daily_data_1, columns=["Date", "CPI (%)"])

# Lưu ra file
output_file_1 = r"C:\Users\Admin\Downloads\NCKH_\TRANSFORM\CPI.csv"


CPI.to_csv(output_file_1, index=False)
```

```python
def modify_volume_column(df, column_name):
    def convert_volume(value):
        value = str(value).strip()
        if 'M' in value:
            return float(value.replace('M', '')) * 1_000_000  # Chuyển đổi giá trị có 'M'
        else:
            return float(value) * 1_000  # Nhân 1000 với giá trị số thông thường
    
    df[column_name] = df[column_name].apply(convert_volume)
    df[column_name] = df[column_name].astype(float)  # Đảm bảo kiểu dữ liệu là float
    return df

VNI = pd.DataFrame(VNI)
VNI = modify_volume_column(VNI, 'Volume')
print(VNI)
```

```python
# Danh sách tên các DataFrame
dataframes = ['CPI', 'CRUDE_OIL', 'DOW_JONES', 'GLOBAL_EPU', 'GOLD', 'GPR',
              'IPI', 'MS', 'NASDAQ', 'SP500', 'OVERNIGHT_INTEREST_RATE', 'US10', 
              'US2', 'US_EPU', 'USD_VND', 'VN10', 'VN2','VNI','VIX']

# Lấy DataFrame CRUDE_OIL làm bảng gốc
merged_df = globals()['CRUDE_OIL']

# Duyệt qua các DataFrame và thực hiện left join
for name in dataframes:
    if name != 'CRUDE_OIL':  # Bỏ qua bảng CRUDE_OIL vì nó đã là DataFrame gốc
        df = globals()[name]  # Lấy DataFrame từ tên biến toàn cục
        merged_df = pd.merge(merged_df, df[['Date', df.columns[1]]], on='Date', how='left')

merged_df = merged_df.merge(VNI, on="Date", how="left")
merged_df = merged_df.merge(FSI, on="Date", how="left")
# Xem kết quả sau khi merge
print(merged_df.head())

```

```python
# Lọc danh sách các cột có kiểu object
object_columns = merged_df.select_dtypes(include=['object']).columns

# Hàm làm sạch và chuyển đổi dữ liệu
def clean_and_convert(series):
    return pd.to_numeric(series.astype(str)
                         .str.replace(',', '', regex=True)  # Xóa dấu phân tách hàng nghìn
                         .str.replace('%', '', regex=True)  # Xóa ký hiệu %
                         .str.strip(),                      # Xóa khoảng trắng dư thừa
                         errors='coerce')  # Nếu lỗi, thay bằng NaN

# Áp dụng cho tất cả các cột dạng object
merged_df[object_columns] = merged_df[object_columns].apply(clean_and_convert)

# Kiểm tra kiểu dữ liệu sau khi chuyển đổi
print(merged_df.dtypes)
```

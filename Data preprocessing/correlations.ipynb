### **SPEARMAN CORRELATION**

##### *Giữa các features với biến mục tiêu*

```python
import pandas as pd
import scipy.stats as stats


# Xác định tên cột mục tiêu và cột Date
target = "VNI_Close price"
date_column = "Date"
# Tạo danh sách lưu kết quả
correlations = []

# Tính Spearman Correlation giữa từng feature và VNI_Close
for feature in df_numeric.columns:
    if feature != target:
        corr, p_val = stats.spearmanr(df_numeric[feature], df_numeric[target])
        correlations.append({"Feature": feature, "Spearman_Correlation": corr, "p-value": p_val})

# Chuyển danh sách thành DataFrame
correlations = pd.DataFrame(correlations)

# Sắp xếp kết quả theo độ tương quan giảm dần
spearman_correlations = correlations.sort_values(by="Spearman_Correlation", ascending=False)
```

```python
spearman_correlations
```

##### *Giữa các features với nhau*

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Tính ma trận tương quan
corr_matrix = df_numeric.corr(method="spearman")

# Vẽ heatmap
plt.figure(figsize=(40, 15))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Spearman Correlation Between Features")
plt.show()
```

### **PEARSON CORRELATIONS**

##### *Giữa các features với biến mục tiêu*

```python
import pandas as pd
import scipy.stats as stats

# Xác định biến mục tiêu và cột Date (nếu có)
target = "VNI_Close price"
date_column = "Date"  # Nếu có cột Date, loại bỏ

# Tạo danh sách lưu kết quả
correlations = []

# Tính Pearson Correlation giữa từng feature và VNI_Close
for feature in df_numeric.columns:
    if feature != target:
        corr, p_val = stats.pearsonr(df_numeric[feature], df_numeric[target])
        correlations.append({"Feature": feature, "Pearson_Correlation": corr, "p-value": p_val})

# Chuyển danh sách thành DataFrame
correlations = pd.DataFrame(correlations)

# Sắp xếp theo độ tương quan giảm dần
pearson_correlations = correlations.sort_values(by="Pearson_Correlation", ascending=False)
```

```python
pearson_correlations
```

##### *Giữa các features với nhau*

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Tính ma trận tương quan
corr_matrix = df_numeric.corr(method="pearson")

# Vẽ heatmap
plt.figure(figsize=(40, 15))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Pearson Correlation Between Features")
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Đọc file dự báo từ tập OOS
df_base = pd.read_csv(r"C:\Users\Admin\Downloads\NCKH_\MODEL\Model training\LSTM_Base\lstm_base_OOS_predictions.csv")
df_advanced = pd.read_csv(r"C:\Users\Admin\Downloads\NCKH_\MODEL\Model training\LSTM_Advanced\lstm_advanced_OOS_predictions.csv")

# Trích xuất giá trị thực tế và dự báo
y_true = df_base["Actual VNI_Close"].values
y_pred_base = df_base["Predicted VNI_Close"].values
y_pred_adv = df_advanced["Predicted VNI_Close"].values

# Hàm CW-test
def cw_test(y_true, y_pred_base, y_pred_adv):
    e_base = y_true - y_pred_base
    e_adv = y_true - y_pred_adv
    f_t = e_base**2 - (e_adv**2 - (y_pred_adv - y_pred_base)**2)
    X = np.ones((len(f_t), 1))  
    model = sm.OLS(f_t, X).fit()
    return model.tvalues[0], model.pvalues[0]

# Thực hiện CW-test giữa LSTM Base và LSTM Advanced
t_stat, p_value = cw_test(y_true, y_pred_base, y_pred_adv)

# In kết quả
print(f"CW-test statistic: {t_stat}")
print(f"p-value: {p_value}")

# Nếu p-value < 0.05, LSTM Advanced tốt hơn đáng kể so với LSTM Base
if p_value < 0.05:
    print("LSTM Advanced có cải thiện đáng kể so với LSTM Base.")
else:
    print("Không có bằng chứng cho thấy LSTM Advanced tốt hơn đáng kể.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Danh sách các tệp CSV và tên mô hình
file_paths = {
    "LSTM Base": r"C:\Users\Admin\Downloads\NCKH_\MODEL\Model training\LSTM_Base\lstm_base_OOS_predictions.csv",
    "LSTM Advanced": r"C:\Users\Admin\Downloads\NCKH_\MODEL\Model training\LSTM_Advanced\lstm_advanced_OOS_predictions.csv",
    "BiLSTM Base": r"C:\Users\Admin\Downloads\NCKH_\MODEL\Model training\BiLSTM_Base\bilstm_base_OOS_predictions.csv",
    "BiLSTM Advanced": r"C:\Users\Admin\Downloads\NCKH_\MODEL\Model training\BILSTM_Advanced\bilstm_advanced_OOS_predictions.csv",
    "Transformer Base": r"C:\Users\Admin\Downloads\NCKH_\MODEL\Model training\Transformer_Base\transformer_base_OOS_predictions.csv",
    "Transformer Advanced": r"C:\Users\Admin\Downloads\NCKH_\MODEL\Model training\Transformer_Advanced\transformer_advanced_OOS_predictions.csv",
}

# DataFrame tổng hợp
df_all = pd.DataFrame()

for model_name, path in file_paths.items():
    df = pd.read_csv(path)

    # Xóa cột thừa nếu có
    if "Unnamed: 0" in df.columns:
        df.drop(columns=["Unnamed: 0"], inplace=True)
    if "Date.1" in df.columns:
        df.drop(columns=["Date.1"], inplace=True)

    # Đổi tên cột cho đồng nhất
    if "Actual VNI_Close" in df.columns:
        df.rename(columns={"Actual VNI_Close": "Actual", "Predicted VNI_Close": model_name}, inplace=True)
    elif "Actual" in df.columns and "Predicted" in df.columns:
        df.rename(columns={"Predicted": model_name}, inplace=True)

    # Chuyển Date sang datetime
    df["Date"] = pd.to_datetime(df["Date"])

    # Nếu df_all đang rỗng, lấy cột Date và Actual
    if df_all.empty:
        df_all = df[["Date", "Actual"]].copy()

    # Merge dữ liệu
    df_all = df_all.merge(df[["Date", model_name]], on="Date", how="left")

# Màu sắc tùy chỉnh cho từng mô hình
colors = {
    "LSTM Base": "blue",
    "LSTM Advanced": "orange",
    "BiLSTM Base": "purple",
    "BiLSTM Advanced": "brown",
    "Transformer Base": "red",
    "Transformer Advanced": "green",  # Transformer Advanced màu xanh lá cây
}

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
plt.plot(df_all["Date"], df_all["Actual"], label="Actual", color="black", linewidth=2)

# Thêm đường dự đoán từ các mô hình với màu sắc tùy chỉnh
for model_name, color in colors.items():
    plt.plot(df_all["Date"], df_all[model_name], label=model_name, linestyle="dashed", color=color)

# Cấu hình biểu đồ
plt.xlabel("Date")
plt.ylabel("VN-Index Price")
plt.title("Actual vs Predicted VN-Index Prices (Multiple Models)")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
